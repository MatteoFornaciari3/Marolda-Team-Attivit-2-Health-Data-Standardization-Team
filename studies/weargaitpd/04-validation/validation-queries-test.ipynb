```python
import pandas as pd

controls_df = pd.read_csv('../01-refinement/clean/controls_first_overview.clean.csv')
pd_df = pd.read_csv('../01-refinement/clean/pd_first_overview.clean.csv')
balance_df = pd.read_csv('../01-refinement/clean/balance.clean.csv')
```

### 1) How many patients have Parkinson disease?

```python

# Tutti i pazienti nel file PD hanno Parkinson
pd_patients_ids = pd_df['patient.subjectid'].unique()

print("Number of patients with Parkinson:", len(pd_patients_ids)) # 85 pazienti
```

### 2) Check PT/OT status (Controls + PD)

```python
controls_df['obssoc.pt_ot_status'] = controls_df['obssoc.pt_ot_status'].astype(str).str.lower().str.strip()
pd_df['obssoc.pt_ot_status'] = pd_df['obssoc.pt_ot_status'].astype(str).str.lower().str.strip()

# Count valori
controls_ptot = controls_df['obssoc.pt_ot_status'].value_counts()
pd_ptot = pd_df['obssoc.pt_ot_status'].value_counts()

print("\nControls PT/OT status:")
print(controls_ptot)

print("\nPD PT/OT status:")
print(pd_ptot)
```

### 3) Patients doing PT/OT (YES)

```python
controls_ptot_yes = controls_df[controls_df['obssoc.pt_ot_status'] == 'yes']['patient.subjectid'].unique()
pd_ptot_yes = pd_df[pd_df['obssoc.pt_ot_status'] == 'yes']['patient.subjectid'].unique()

print("Controls doing PT/OT:", len(controls_ptot_yes))
print("PD doing PT/OT:", len(pd_ptot_yes))
```

### 4) Frequency PT/OT

```python
controls_df['obssoc.frequency_pt_ot'] = controls_df['obssoc.frequency_pt_ot'].astype(str).str.lower().str.strip()
pd_df['obssoc.frequency_pt_ot'] = pd_df['obssoc.frequency_pt_ot'].astype(str).str.lower().str.strip()

pd_frequency = pd_df['obssoc.frequency_pt_ot'].value_counts()
controls_frequency = controls_df['obssoc.frequency_pt_ot'].value_counts()

print("PD PT/OT frequency:")
print(pd_frequency)

print("\nControls PT/OT frequency:")
print(controls_frequency)
```
